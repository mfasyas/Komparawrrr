```python
import markov_clustering as mc
import networkx as nx
import numpy as np

# Example: create adjacency matrix and run MCL
matrix = np.array([[1, 1, 0, 0],
                   [1, 1, 0, 0],
                   [0, 0, 1, 1],
                   [0, 0, 1, 1]])

result = mc.run_mcl(matrix)           # Run MCL
clusters = mc.get_clusters(result)    # Get clusters
print(clusters)  # e.g., [(0, 1), (2, 3)]
```

```python
clusters
```

```python
G = nx.from_numpy_array(matrix)

# Assign cluster label to each node
cluster_dict = {}
for cluster_id, cluster in enumerate(clusters):
    for node in cluster:
        cluster_dict[node] = cluster_id

# Add cluster info as node attributes
nx.set_node_attributes(G, cluster_dict, 'cluster')
```

```python
import matplotlib.pyplot as plt

# Use a spring layout (or any layout you prefer)
pos = nx.spring_layout(G)

# Create color map
colors = [cluster_dict[node] for node in G.nodes]

# Draw the graph
nx.draw(G, pos, node_color=colors, with_labels=True, cmap=plt.cm.Set3, node_size=800)

plt.show()
```

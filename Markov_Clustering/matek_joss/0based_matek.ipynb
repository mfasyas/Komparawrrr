```python
import pandas as pd
```

```python
df = pd.read_csv("F:\\Coding\\Komparawrrr\\Markov_Clustering\\matek_joss\\gugel_scholar.csv")
print(df.head())
```

```python
df = df[["Author", "Co Author"]]
print(df.head())
```

```python
import networkx as nx
import numpy as np
```

```python
G = nx.from_pandas_edgelist(df,source = 'Author', target = "Co Author")
```

```python
# draw the graph
import matplotlib.pyplot as plt

posi = nx.spring_layout(G, k = 0.2, iterations = 50, seed = 42)

plt.figure(figsize=(18, 14))
nx.draw(
    G,
    pos = posi,
    with_labels=True,
    node_color='red',
    edge_color='black',
    node_size=1000,
    font_size=8
)
plt.title("Graph of LastFM User Network")
plt.show()
```

```python
adj_matek = nx.to_numpy_array(G, dtype = np.float32)

print("Matrix adjacency:\n", adj_matek)
```

```python
import markov_clustering as mc
result = mc.run_mcl(adj_matek, inflation=2)
clusters = mc.get_clusters(result)
```
